### Exercise 25

Setting the scene: build a beam line of 6 FODO cells (e.g., L_cell = 4 m) with a phase advance of 60 degrees/cell (using 3x3 thin quadrupoles just introduced). 
Add a sector bending magnet (3x3 version) with length 1 m and bending angle $\phi=10$ degrees in the center of each drift. 
At this point, it is not too important where you start your FODO cell.

Verify nothing changes with respect to the same beamline made of 2x2 matrices.

- **Note:** You may have to play with the quadrupole focal length to make the phase advance close to 60 degrees. You should have done this in earlier exercises. 

#### SOLUTION

```python
# building up the 3x3 case
n_cells = 6
f = 2.26085
l_cell = 4
l_dipole = 1
l_drift = (l_cell - l_dipole * 2) / 4
phi = 10 * np.pi / 180
fodo_cell = (
    Q3(f)
    + 3 * D3(l_drift / 3)
    + B3(phi, l_dipole)
    + 3 * D3(l_drift / 3)
    + Q3(-f)
    + 3 * D3(l_drift / 3)
    + B3(phi, l_dipole)
    + 3 * D3(l_drift / 3)
)
fodo_lattice = n_cells * fodo_cell
phase, beta, alpha, gamma = twiss(fodo_cell)
tune = n_cells * phase
print(
    f"""
The 3x3 phase per cell is: {phase}
The 3x3 beta is: {beta} m
The 3x3 alpha is: {alpha} 
The 3x3 gamma is: {gamma}
"""
)
```

```python
# repeating for 2x2 case
fodo_cell_2x2 = (
    Q(f)
    + 3 * D(l_drift / 3)
    + B(phi, l_dipole)
    + 3 * D(l_drift / 3)
    + Q(-f)
    + 3 * D(l_drift / 3)
    + B(phi, l_dipole)
    + 3 * D(l_drift / 3)
)
phase, beta, alpha, gamma = twiss(fodo_cell_2x2)
print(
    f"""
The 2x2 phase per cell is: {phase}
The 2x2 beta is: {beta} m
The 2x2 alpha is: {alpha} 
The 2x2 gamma is: {gamma}
"""
)
```

### Exercise 29

Determine the periodic dispersion at the start of the cell. Then plot the dispersion in the cell.

#### SOLUTION

For that we need to find the *real* periodic solution of the 3x3 one turn matrix

```python
OTM = getEquivalentElement(beamline)
eigenvalues, eigenvectors = np.linalg.eig(OTM[0]["matrix"])
print(eigenvalues)
print(eigenvectors[:, 2])
```

Note that the third eigenvalue is real, i.e. is associated to a real eigenvector, which is our solution:

```python
dispersion = np.real(eigenvectors[:, 2] / eigenvectors[2, 2])
print(dispersion)
```

Sanity check: one can see that indeeed is a periodic solution!

```python
OTM[0]["matrix"] @ dispersion
```

And track it along our lattice:

```python
# need to make dispersion array compatible with our transportParticles function
dispersion = np.reshape(dispersion, (3, 1))
transported_particles = transportParticles(dispersion, beamline)

plt.plot(transported_particles["s"], transported_particles["x"], "o-b", lw=3)
plt.grid(True)
plt.xlabel("s [m]")
plt.ylabel("Dx [m]", color="b")
plt.tick_params(axis="y", labelcolor="b")
```

### Exercise 4.1Ph

Let's consider a simple FODO lattice starting from the middle of a drift, for example $L_{FODO} = 4$ m and 60deg phase advance (thin quadrupoles are OK) and n_cells = 100.

1. Generate a **matched** beam distribution (e.g. 1000 particles) using the `beam()` function from `tracking_library_advanced.py`, track them once through the FODO cell, plot the final $(x\, x')$ phase space, and compare it with the initial particle coordinates: did the emittance change?

2. Now assume to have several (e.g. 10) `beamlets` of particles which have slightly different momentum (e.g. -1% <= $\delta p/p_0$  <= 1\%). Track those bemlets along the same beamline, but adjusting the focal length of the quadrupoles to take into account the different particle momentum.

    **Hint:** Remember that the relationship between focal length $f$ and integrated quadrupoles strenght $K1 L$ is:

    \begin{equation}
    f = \frac{1}{K1 \times L}
    \end{equation}

    and that $K1$ depends on the particle momentum $p$ as:

    \begin{equation}
    K1(p) = \frac{q}{p}g
    \end{equation}

    where $g$ is the quadrupole gradient, $q$ is the particle charge. Hence, by Taylor expansion:

    \begin{equation}
    f \left(\frac{\delta p}{p_0} \right) \approx f_{p_0} + f_{p_0} \left(\frac{\delta p}{p_0} \right) + \ldots 
    \qquad \text{or} \qquad 
    K1 \left(\frac{\delta p}{p_0} \right) \approx K1_{p_0} - K1_{p_0} \left(\frac{\delta p}{p_0} \right) + \ldots
    \end{equation}

    Looking by eye at the final phase space of the sovraposition of all beamlets, would you say the emittance has changed?

3. Repeat the previous point, but this time start from an un-matched distribution, e.g. flip the sign of Twiss parameter $alpha$: what does it happen to the final emittance? 

### SOLUTION 1

```python
from tracking_library_advanced_solution import beam

# parameters for the FODO cell and lattice
L_2 = 4
L = L_2 / 2
desired_mu = 60 * np.pi / 180
f = L_2 / (4 * np.sin(desired_mu / 2))
n_cells = 100


# generator of FODO cells
def generate_fodo_lattice(f, L, n_cells=1):
    return n_cells*(4 * D(L / 8) + Q(f) + 8 * D(L / 8) + Q(-f) + 4 * D(L / 8))

# generate un perturbed fodo cell
fodo_lattice_base = generate_fodo_lattice(f, L, n_cells)
# compute matched parameters
(phase, beta, alpha, gamma) = twiss(fodo_lattice_base)

# generate initial (matched) beam (assuming emittance=1, n_particles=1000)
initial_matched_coordinates = beam(1, beta, alpha, 1000)

# track it along the beamline 
tracked_matched_coordinates = transportParticles(initial_matched_coordinates, fodo_lattice_base)

# plot them in a plot
plt.figure()
plt.scatter(tracked_matched_coordinates["x"][-1,:], tracked_matched_coordinates["xp"][-1,:])
plt.xlabel("x [mm]")
plt.ylabel("x' [mrad]")
plt.grid(True)

# plot initial distribution
plt.scatter(initial_matched_coordinates[0,:], initial_matched_coordinates[1,:], color='k', marker='x')
```

- **The emittance seems to be constant**: no change on the surface of phase-space being occupied

### SOLUTION 2

```python
# prepare a plot
plt.figure()
plt.xlabel("x [mm]")
plt.ylabel("x' [mrad]")
plt.grid(True)

# range of dp/p
dp_p_list = np.linspace(-0.01, 0.01, 11)
for dp_p in dp_p_list:
    # perturbed fodo
    _fodo_cell = generate_fodo_lattice(f*(1+dp_p), L, n_cells)
    # track particles
    _tracked = transportParticles(initial_matched_coordinates, _fodo_cell)
    # add them to the plot
    plt.scatter(_tracked["x"][-1,:], _tracked["xp"][-1,:])

# plot initial distribution
plt.scatter(initial_matched_coordinates[0,:], initial_matched_coordinates[1,:], color='k', marker='x')

```

### SOLUTION 3

```python
# generate initial (matched) beam (assuming emittance=1, n_particles=1000)
initial_unmatched_coordinates = beam(1, beta, -alpha, 1000)

# prepare figure
plt.figure()
plt.xlabel("x [mm]")
plt.ylabel("x' [mrad]")
plt.grid(True)

# range of dp/p
dp_p_list = np.linspace(-0.01, 0.01, 11)
for dp_p in dp_p_list:
    # perturbed fodo
    _fodo_cell = generate_fodo_lattice(f*(1+dp_p), L, n_cells)
    # track particles
    _tracked = transportParticles(initial_unmatched_coordinates, _fodo_cell)
    # add them to the plot
    plt.scatter(_tracked["x"][-1,:], _tracked["xp"][-1,:])

# plot initial distribution
plt.scatter(initial_unmatched_coordinates[0,:], initial_unmatched_coordinates[1,:], color='k', marker='x')
```

- **Note** that emittance is now far from being constant! **You are observing filamentation** due to chromaticity!
